```r
library(DropletUtils)
```

```r
rna <- read10xCounts("438-21-raw_feature_bc_matrix.h5")
rna
```

```r
rna_matrix <- counts(rna)
rownames(rna_matrix) <- rownames(rna)
colnames(rna_matrix) <- colData(rna)$Barcode

rna_matrix
```

```r
# Convert to sparse numeric matrix
rna_matrix <- as(rna_matrix, "sparseMatrix")
rna_matrix
```

```r
write10xCounts("rna_converted", rna_matrix, type = "sparse")
```

```r
hto <- read.csv("438_21_raw_HTO.csv", row.names = 1)
hto_matrix <- as.matrix(hto)
hto_matrix <- as(hto_matrix, "sparseMatrix")
hto_matrix
```

```r
write10xCounts("hto_converted", hto_matrix, type = "sparse")
```
